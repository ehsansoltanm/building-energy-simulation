# Select Best Solution from Pareto Front

This notebook loads the Pareto optimal solutions identified in the previous step (`pareto_optimal_solutions.csv`) and selects a single "best" solution based on a balanced approach. It normalizes the objective values and chooses the solution with the minimum average normalized score (or minimum distance from the ideal point).

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import os
```

## Define File Paths and Objectives

```python
# Input file containing Pareto optimal solutions
PARETO_SOLUTIONS_FILE = "/home/ubuntu/pareto_optimal_solutions.csv"

# Output file for the selected best solution details
BEST_SOLUTION_FILE = "/home/ubuntu/selected_solution.txt"

# Define the objectives (ensure these match the columns in the CSV)
objectives = ["Electricity:Facility", "DistrictHeating:Facility", "DistrictCooling:Facility"]
```

## Load Pareto Solutions

```python
try:
    df_pareto = pd.read_csv(PARETO_SOLUTIONS_FILE)
    print(f"Loaded {len(df_pareto)} Pareto optimal solutions from {PARETO_SOLUTIONS_FILE}")
except FileNotFoundError:
    print(f"Error: Pareto solutions file not found at {PARETO_SOLUTIONS_FILE}")
    df_pareto = None # Prevent further execution

# Verify objectives exist
if df_pareto is not None:
    missing_objectives = [obj for obj in objectives if obj not in df_pareto.columns]
    if missing_objectives:
        print(f"Error: The following objective columns are missing in {PARETO_SOLUTIONS_FILE}: {missing_objectives}")
        df_pareto = None
```

## Normalize Objective Values

Normalize the objective values to a 0-1 scale, where 0 represents the best possible value (minimum) and 1 represents the worst (maximum) among the Pareto set for that objective.

```python
if df_pareto is not None:
    scaler = MinMaxScaler()
    df_pareto_normalized = pd.DataFrame(scaler.fit_transform(df_pareto[objectives]), 
                                        columns=[f"{obj}_norm" for obj in objectives],
                                        index=df_pareto.index)
    
    # Combine normalized values with the original DataFrame
    df_pareto = pd.concat([df_pareto, df_pareto_normalized], axis=1)
    
    print("
Normalized Objective Values (First 5):")
    print(df_pareto[[f"{obj}_norm" for obj in objectives]].head())
else:
    print("Skipping normalization due to data loading errors.")
```

## Calculate Score and Select Best Solution

Calculate a simple score for each solution. Here, we use the average of the normalized objective values. The solution with the lowest score is considered the most balanced.

```python
if df_pareto is not None:
    # Calculate a simple score: average normalized objective value (lower is better)
    df_pareto["Normalized_Score"] = df_pareto[[f"{obj}_norm" for obj in objectives]].mean(axis=1)
    
    # Alternative: Euclidean distance from the ideal point (0, 0, 0)
    # df_pareto["Distance_From_Ideal"] = np.sqrt((df_pareto[[f"{obj}_norm" for obj in objectives]]**2).sum(axis=1))
    
    # Select the solution with the minimum score
    best_solution_index = df_pareto["Normalized_Score"].idxmin()
    best_solution = df_pareto.loc[best_solution_index]
    
    print("
Selected Best Solution (Based on Minimum Average Normalized Score):")
    print(best_solution)
else:
    print("Skipping best solution selection due to previous errors.")
    best_solution = None
```

## Save Best Solution Details

```python
if best_solution is not None:
    try:
        with open(BEST_SOLUTION_FILE, "w") as f:
            f.write("Selected Best Solution (Balanced Approach - Min Avg Normalized Score):\n")
            f.write("==================================================================\n")
            # Write original parameters and objective values
            f.write(best_solution.to_string())
        print(f"
Best solution details saved to {BEST_SOLUTION_FILE}")
    except IOError as e:
        print(f"Error saving best solution details to {BEST_SOLUTION_FILE}: {e}")
else:
    print("Skipping saving best solution details.")
```
