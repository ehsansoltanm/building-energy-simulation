# Energy Signature Analysis

This notebook analyzes the relationship between outdoor temperature and the building's heating and cooling energy consumption using data from an EnergyPlus simulation (`eplusout.csv`) and the corresponding weather file (`.epw`). It generates energy signature plots for heating and cooling.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys

# Ensure plots appear inline
%matplotlib inline
```

## Configuration

```python
# Define temperature thresholds for balance points
HEATING_BALANCE_POINT = 15.0
COOLING_BALANCE_POINT = 20.0

# Define file paths (adjust if necessary)
EPW_FILE = "/home/ubuntu/upload/Torino_IT-hour.epw"
SIM_FILE = "/home/ubuntu/upload/eplusout.csv"
OUTPUT_DIR = "/home/ubuntu/output_energy_signature"

# Create output directory if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"Input EPW file: {EPW_FILE}")
print(f"Input Simulation file: {SIM_FILE}")
print(f"Output directory: {OUTPUT_DIR}")
```

## Helper Function: Load and Preprocess Data

```python
def load_data(epw_path, sim_path):
    """Loads and preprocesses weather and simulation data."""
    # Load Weather Data
    try:
        epw = pd.read_csv(epw_path, skiprows=8, header=None)
        epw.columns = ["Year", "Month", "Day", "Hour", "Minute", "DataSource", "DryBulb", "DewPoint", "RH", "AtmosPressure", 
                       "ExtGlobHorRad", "ExtDirNormRad", "ExtDifHorRad", "GlobalHorRad", "DirectNormRad", 
                       "DiffuseHorRad", "InfraSky", "WindDir", "WindSpd", "TotalSkyCover", "OpaqueSkyCover", 
                       "Visibility", "CeilingHeight", "PresWeatherObs", "PresWeatherCodes", "PrecipWater", "AerosolOptDepth",
                       "SnowDepth", "DaysSinceSnow", "Albedo", "LiquidPrecip", "RainRate", "RainDuration", "SnowRate", "SnowDuration"]
        weather_df = epw[["DryBulb"]].copy()
        weather_df.columns = ["outdoor_temp_c"]
        print(f"Successfully loaded weather data from {epw_path}")
    except FileNotFoundError:
        print(f"Error: EPW file not found at {epw_path}")
        return None
    except Exception as e:
        print(f"Error loading EPW file {epw_path}: {e}")
        return None

    # Load Simulation Data
    try:
        sim_df = pd.read_csv(sim_path, low_memory=False)
        # Find heating and cooling columns dynamically
        heating_col = next((c for c in sim_df.columns if "DistrictHeating:Facility" in c), None)
        cooling_col = next((c for c in sim_df.columns if "DistrictCooling:Facility" in c), None)
        
        if not heating_col or not cooling_col:
            print("Warning: Standard Heating/Cooling columns not found. Trying exact names...")
            heating_col = "DistrictHeating:Facility"
            cooling_col = "DistrictCooling:Facility"
            if heating_col not in sim_df.columns or cooling_col not in sim_df.columns:
                 print("Error: Could not find required heating/cooling columns even with exact names.")
                 return None
            else:
                 print("Found columns using exact names: '{heating_col}', '{cooling_col}'")
        else:
             print(f"Found columns dynamically: '{heating_col}', '{cooling_col}'")
            
        sim_df_hourly = sim_df[[heating_col, cooling_col]].copy()
        sim_df_hourly.columns = ["heating_j", "cooling_j"]
        print(f"Successfully loaded simulation data from {sim_path}")
        
    except FileNotFoundError:
        print(f"Error: Simulation file not found at {sim_path}")
        return None
    except Exception as e:
        print(f"Error loading simulation file {sim_path}: {e}")
        return None

    # Combine DataFrames
    print("Combining weather and simulation data...")
    weather_df_reset = weather_df.reset_index(drop=True)
    sim_df_hourly_reset = sim_df_hourly.reset_index(drop=True)
    min_length = min(len(weather_df_reset), len(sim_df_hourly_reset))
    
    if min_length < len(weather_df_reset) or min_length < len(sim_df_hourly_reset):
        print(f"Warning: Weather ({len(weather_df_reset)} rows) and Simulation ({len(sim_df_hourly_reset)} rows) data have different lengths. Truncating to {min_length} rows.")
        
    full_df = pd.concat([weather_df_reset.iloc[:min_length], sim_df_hourly_reset.iloc[:min_length]], axis=1)
    
    # Drop rows with NaN values that might result from merging or issues in source files
    initial_rows = len(full_df)
    full_df = full_df.dropna()
    final_rows = len(full_df)
    if initial_rows > final_rows:
        print(f"Dropped {initial_rows - final_rows} rows containing NaN values.")
        
    print(f"Combined DataFrame created with {final_rows} rows.")
    return full_df
```

## Load Data

```python
data = load_data(EPW_FILE, SIM_FILE)

if data is not None:
    print("
Data loaded successfully. First 5 rows:")
    print(data.head())
else:
    print("
Failed to load data. Cannot proceed with analysis.")
```

## Generate Heating Energy Signature Plot

```python
if data is not None:
    print("
Generating Heating Energy Signature plot...")
    heating_data = data[data["heating_j"] > 1e-6] # Filter out zero or negligible heating
    
    if not heating_data.empty:
        plt.figure(figsize=(10, 6))
        plt.scatter(heating_data["outdoor_temp_c"], heating_data["heating_j"], alpha=0.5, s=10) # Smaller points
        plt.axvline(HEATING_BALANCE_POINT, color='r', linestyle='--', label=f"Balance Point ({HEATING_BALANCE_POINT}째C)")
        plt.title("Heating Energy Signature")
        plt.xlabel("Outdoor Temperature (째C)")
        plt.ylabel("Heating Energy (J)")
        plt.legend()
        plt.grid(True)
        
        heating_plot_path = os.path.join(OUTPUT_DIR, "energy_signature_heating.png")
        try:
            plt.savefig(heating_plot_path)
            print(f"Heating signature plot saved to {heating_plot_path}")
            plt.show() # Display plot in notebook
        except Exception as e:
            print(f"Error saving heating plot: {e}")
        plt.close() # Close the plot figure
    else:
        print("No significant heating data found to plot.")
else:
    print("Skipping heating plot generation due to data loading errors.")
```

## Generate Cooling Energy Signature Plot

```python
if data is not None:
    print("
Generating Cooling Energy Signature plot...")
    cooling_data = data[data["cooling_j"] > 1e-6] # Filter out zero or negligible cooling
    
    if not cooling_data.empty:
        plt.figure(figsize=(10, 6))
        plt.scatter(cooling_data["outdoor_temp_c"], cooling_data["cooling_j"], alpha=0.5, s=10) # Smaller points
        plt.axvline(COOLING_BALANCE_POINT, color='r', linestyle='--', label=f"Balance Point ({COOLING_BALANCE_POINT}째C)")
        plt.title("Cooling Energy Signature")
        plt.xlabel("Outdoor Temperature (째C)")
        plt.ylabel("Cooling Energy (J)")
        plt.legend()
        plt.grid(True)
        
        cooling_plot_path = os.path.join(OUTPUT_DIR, "energy_signature_cooling.png")
        try:
            plt.savefig(cooling_plot_path)
            print(f"Cooling signature plot saved to {cooling_plot_path}")
            plt.show() # Display plot in notebook
        except Exception as e:
            print(f"Error saving cooling plot: {e}")
        plt.close() # Close the plot figure
    else:
        print("No significant cooling data found to plot.")
else:
    print("Skipping cooling plot generation due to data loading errors.")
```
