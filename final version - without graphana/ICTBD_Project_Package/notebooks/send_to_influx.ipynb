```python
import pandas as pd
from influxdb_client import InfluxDBClient, Point, WriteOptions
from datetime import datetime

url = "http://host.docker.internal:8086"
token = "43s4wbfW--cqzkK11gEATVvc2yPlxyUU-_pgjHpqJLGwgvZd0peN4BzoWjkfhcJ7pPqXTy6PX1D6dcPW9bZfVQ=="
org = "building_lab"
bucket = "energy_data"

client = InfluxDBClient(url=url, token=token, org=org)
write_api = client.write_api(write_options=WriteOptions(batch_size=500))

df = pd.read_csv("eplusout.csv")
df["DateTime"] = pd.to_datetime(df["Date/Time"], errors="coerce")
df = df.dropna(subset=["DateTime"])

columns = [
    "DateTime",
    "Environment:Site Outdoor Air Drybulb Temperature [C](TimeStep)",
    "Environment:Site Wind Speed [m/s](TimeStep)",
    "Environment:Site Direct Solar Radiation Rate per Area [W/m2](TimeStep)"
]

df = df[columns]

for _, row in df.iterrows():
    point = (
        Point("weather_data")
        .field("temperature", float(row[columns[1]]))
        .field("wind_speed", float(row[columns[2]]))
        .field("solar_radiation", float(row[columns[3]]))
        .time(row["DateTime"])
    )
    write_api.write(bucket=bucket, org=org, record=point)

client.close()
print("✅ داده‌ها ارسال شدند.")
```
