```python
from influxdb_client import InfluxDBClient

url = "http://influxdb:8086"
token = "43s4wbfW--cqzkK11gEATVvc2yPlxyUU-_pgjHpqJLGwgvZd0peN4BzoWjkfhcJ7pPqXTy6PX1D6dcPW9bZfVQ=="
org = "building_lab"
bucket = "energy_data"

client = InfluxDBClient(url=url, token=token, org=org)

query = f'''
from(bucket: "{bucket}")
  |> range(start: -1h)
  |> filter(fn: (r) => r["_measurement"] == "energy_usage")
  |> filter(fn: (r) => r["_field"] == "energy_joule")
'''

tables = client.query_api().query(query=query)

for table in tables:
    for record in table.records:
        print(f"time: {record.get_time()}, value: {record.get_value()}, tags: {record.values}")
```
